```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = pd.Series([24.23,25.53,25.41,24.14,29.62,28.25,25.81,24.39,40.26,32.95,91.36,25.99,39.42,26.71,35.00])
company = ['Allied Signal','Bankers Trust','General','Mills','ITT Industries','J.P.Morgan & Co.','Lehman Brothers'
           ,'Marriott','MCI','Merrill Lynch','Microsoft','Morgan Stanley','Sun Microsystems','Travelers'
           ,'US Airways','Warner-Lambert']
```

```python
plt.figure(figsize=(10,10),facecolor="lightgreen")
plt.pie(x,labels=company,autopct="%2.2f%%")
plt.xlabel("Measure X")
plt.ylabel("Name of Company")
plt.show()
```

```python
x.mean()
```

```python
x.std()
```

```python
x.var()
```

```python
q1 = x.quantile(0.25)
q1
```

```python
q3 = x.quantile(0.75)
q3
```

```python
iqr = q3 - q1
iqr
```

```python
lower_limit = q1 - (1.5 * iqr)
lower_limit
```

```python
upper_limit = q3 + (1.5 * iqr)
upper_limit
```

```python
x[(x < lower_limit) | (x > upper_limit)]
```
